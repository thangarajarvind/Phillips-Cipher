```python
import numpy as np
import time
```

```python
#MATRIX FORMATION
R=5
C=5
least=input("Enter the least occuring character :")
least=least.lower()
print("Enter the entries in a single line (separated by space): ")
x = list(map(str, input().split()))

#Matrix formation
print("The entered matrix is:")
m1= np.array(x).reshape(R, C)
print(m1)

#Duplicating matrix
m2=np.array(x).reshape(R, C)
m3=np.array(x).reshape(R, C)
m4=np.array(x).reshape(R, C)
m5=np.array(x).reshape(R, C)

#Interchanging Rows
m2[[0, 1]] = m2[[1, 0]]
m3[[0, 2]] = m3[[2, 0]]
m4[[0, 3]] = m4[[3, 0]]
m5[[0, 4]] = m5[[4, 0]]

#Interchanging rows with respect to matrix 5
m6=np.copy(m5)
m7=np.copy(m6)
m8=np.copy(m7)
m6[[0, 1]] = m6[[1, 0]]
m7[[0, 2]] = m7[[2, 0]]
m8[[0, 3]] = m8[[3, 0]]
```

```python
#User interface
print("Press 1 for Encryption")
print("Press 2 for Decryption")
z=int(input())
while((z!=1) & (z!=2)):
  print("Enter valid input:")
  z=int(input())
```

```python
#ENCRYPTION
if(z==1):
  #Plain text processing
  ptin=input("Enter the plain text:")

  #Punctuation removal
  pt = ""
  for char in ptin:
    if char.isalpha():
      pt += char
  pt=pt.lower()
  pt=pt.replace(least, "")  #Remove the least occuring character
  #print(pt)

  start = time.time() #Initiates the time

  #Block seperation
  x = pt
  block, blk_size = len(x), 5
  blocks=[x[i:i+blk_size] for i in range(0, block, blk_size)]
  #print("Blocks before padding:")
  #print(blocks)

  #Padding
  if(len(x)%5!=0):
    pad=5-(len(x)%5)
    l="a"
    for i in range(0,pad-1):
      l=l+"a"
    blocks[len(blocks)-1]=blocks[len(blocks)-1]+l
  #print("Blocks after padding:")
  #print(blocks)

  enc=""
  for i in range (0,len(blocks)):
    #To assign matrix m1 to m8
    if i%8==0:
      b=m1
    elif i%8==1:
      b=m2
    elif i%8==2:
      b=m3
    elif i%8==3:
      b=m4
    elif i%8==4:
      b=m5
    elif i%8==5:
      b=m6
    elif i%8==6:
      b=m7
    elif i%8==7:
      b=m8

    for j in range (0,5):
      a=np.where(b==blocks[i][j]) #Mapping letters
      a=list(zip(a[0],a[1])) #Position of letter
      str1=' '.join(map(str,a)) #Converting to string
      res= str1[1:len(str1)-1] #Remval of()
      res=(res.split(", "))
      k=int(res[0])+1 #Offset shift
      l=int(res[1])+1
      if(k>4): #Matrix overflow
        k=k%5
      if(l>4):
        l=l%5
      #print(k,l)
      e=b[k][l] #Finding Cipher alphabet
      enc+=e #Concatenation

  print("\n")
  print("Cipher Text:")
  print(enc.upper())
  end = time.time() #Stops timer
  t=end-start #Time calculation

  file1 = open("myfile.txt","w")
  file1.writelines(enc.upper())
  file1.close()

  print("Time taken for encryption:%.5f" %t)
```

```python
#DECRYPTION
if(z==2):
  #Cipher text processing
  ct=input("Enter the cipher text:")
  ct=ct.lower() #Converting to lower case
  start = time.time() #Initiates the time

  #Block seperation
  x = ct
  block, blk_size = len(x), 5
  blocks=[x[i:i+blk_size] for i in range(0, block, blk_size)]
  #print("Blocks before padding:")
  #print(blocks)
  #Padding
  if(len(x)%5!=0):
    pad=5-(len(x)%5)
    l="A"
    for i in range(0,pad-1):
      l=l+"A"
    blocks[len(blocks)-1]=blocks[len(blocks)-1]+l
  #print("Blocks after padding:")
  #print(blocks)

  dec=""
  for i in range (0,len(blocks)):
    #To assign matrix m1 to m8
    if i%8==0:
      b=m1
    elif i%8==1:
      b=m2
    elif i%8==2:
      b=m3
    elif i%8==3:
      b=m4
    elif i%8==4:
      b=m5
    elif i%8==5:
      b=m6
    elif i%8==6:
      b=m7
    elif i%8==7:
      b=m8

    for j in range (0,5):
      a=np.where(b==blocks[i][j]) #Mapping letters
      a=list(zip(a[0],a[1])) #Position of letter
      #print(a)
      str1=' '.join(map(str,a)) #Converting to string
      res= str1[1:len(str1)-1] #Remval of()
      #print(str1)
      res=(res.split(", "))
      k=int(res[0])-1 #Offset shift
      l=int(res[1])-1
      if(k<0): #Matrix overflow
        k=k%5
      if(l<0):
        l=l%5
      #print(k,l)
      d=b[k][l] #Finding Plain alphabet
      dec+=d #Concatenation
  print("\n")
  print("Plain Text:")
  print(dec.lower())
  end = time.time() #Stops timer
  t=end-start #Time calculation
  print("Time taken for encryption:%.5f" %t)
```
